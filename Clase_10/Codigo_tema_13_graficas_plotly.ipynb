# Graficando con Matplotlib

```python
import numpy as np
t = np.arange(0, 10*np.pi, 0.001)
f = np.cos(t)

import matplotlib.pyplot as plt
plt.plot(t,f, "blue")
plt.title("Gráfica coseno")
```

# Graficando con Plotly

```python
import plotly.graph_objects as go

fig = go.Figure(
                data=[go.Scatter(x=t, y = f, mode="lines", line=dict(width=2, color="blue"))],
                layout=go.Layout(template="simple_white", plot_bgcolor = "#E8FFD8", paper_bgcolor = "#E8FFD8", 
                                 title = {"text": "Gráfica coseno","x" : 0.5}, 
                                 titlefont = {"family": "Arial", "size": 30})   
               )
  
fig.show()
```

# Animando gráfica

```python
t2 = np.arange(0, 1, 0.001)

fig2 = go.Figure()
trace1 = go.Scatter(mode="lines",visible=True,line=dict(color="Blue", width=2),name="Curva",x=t2,y=np.cos(t2))
fig2.add_trace(trace1)
frames = []
for i in range(2,51,1):
  frame = go.Frame(data= [go.Scatter( mode="lines",visible=True,line=dict(color="Blue", width=2),name="Curva " + str(i),x=t2,y=i*np.cos(i*t2))])
  frames.append(frame)


 

fig2.update_layout(
                template= "simple_white", plot_bgcolor = "#E8FFD8", paper_bgcolor = "#E8FFD8",
                title = {"text": "Animación coseno","x" : 0.5}, 
                titlefont = {"family": "Arial", "size": 30},
                yaxis=dict(range=[-51, 51], autorange=False, zeroline=False), showlegend = False, 
                updatemenus=[dict(type="buttons", showactive = False, bordercolor = "black", bgcolor = "#D2FFE8", 
                                  buttons=[dict(label="Play", method="animate", 
                                           args=[None, {"frame": {"duration": 30, "redraw": False},
                                                       "mode": "immediate", "transition": {"duration": 0}
                                                       }
                                                 ]
                                                )
                                          ]
                                  )
                            ]
                  )

fig2.frames=frames
fig2.show()
```
