# Primer gráfico

```python
import matplotlib.pyplot as plt
```

```python
N = 100
X = []
Y = []
for i in range(1, N+1):
  X.append(i)
  Y.append(i)
```

```python
print(X)
print(Y)
```

```python
plt.plot(X, Y)
plt.show()
```

# Agregando ejes y un título al gráfico

```python
plt.plot(X, Y, label = 'Gráfica 1')
plt.title('Mi primer gráfico')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(X, Y, s=None, c=None, marker=None,  alpha=None)
plt.show()
```

# Segundo gráfico

```python
from math import *
import numpy as np
```

## Funciones a graficar

```python
#0.8sin(200 - 2πx)
#1.1cos(250 - πx)
```

```python
X = np.arange(-10,10.01,0.01)
Y_sin = []
Y_cos = []
x = -10
for x in X:
  Y_sin.append( 0.8*sin(200 - 2*x*pi) )
  Y_cos.append( 1.1*cos(250 - pi*x) )
```

```python
print(X)
print(Y_sin)
print(Y_cos)
```

```python
plt.plot(X,Y_sin)
```

```python
plt.plot(X,Y_cos)
```

```python
plt.plot(X,Y_sin, label = 'y = 0.8sin(200 - 2πx')
plt.plot(X,Y_cos, label = 'y = 1.1cos(250 - πx)')
plt.legend()
```

```python
Y_sup = []
for n in range(len(X)):
  Y_sup.append(Y_sin[n] + Y_cos[n])
#print(Y_sup)
```

```python
plt.plot(X, Y_sup, label = 'Pulsos de ondas')
plt.legend()
```

```python
#plt.figure(figsize=(20, 10))
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20,5))
fig.suptitle('Usando subplots')
ax1.plot(X, Y_sin, 'b')
ax2.plot(X, Y_cos, 'r')
ax3.plot(X, Y_sup, 'k')
ax3.scatter(X, Y_sup, marker= '*', c ='tab:orange', s=50, alpha=0.5)
ax1.set_ylabel('amplitud')
ax2.set_xlabel('tiempo')
```

# Creando datos artificiales

Estos pueden usarlos para su tarea (vienen hasta el final) o bien crear sus propios datos con funciones que a ustedes se les ocurran.

```python
from sklearn.datasets import make_blobs
x, y = make_blobs(n_samples=500, centers=3, cluster_std=.90, random_state=3)
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
from sklearn.datasets import make_moons
x, y = make_moons(n_samples=500, noise=.2, random_state=3)
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
from sklearn.datasets import make_circles
x, y = make_circles(n_samples=500, noise=.1, random_state=3, factor = .2)
plt.scatter(x[:,0], x[:,1], c=y)
```
